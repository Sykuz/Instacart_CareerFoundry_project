# 04. Loyalty Flag for Existing Customers

```python
# Creating a new column
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check Columns
df_ords_prods_merge.head(5)
```

```python
# Assign Loyalty Flag based on max_orders
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Check Loyalty Flag
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag in the dataframe using a column specific check
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number', 'max_order']].head(60)
```

```python
# Drop duplicate user to properly check flags
df_loyalty_check = df_ords_prods_merge.drop_duplicates(subset = 'user_id')
```

```python
#View more rows when using head
pd.set_option('display.max_rows', 100)
```

```python
# Re-check the loyalty flag in the drop duplicate dataframe using a column specific check
df_loyalty_check[['user_id', 'loyalty_flag', 'max_order']].head(100)
```

```python
# Check Loyalty Flag count
df_loyalty_check['loyalty_flag'].value_counts(dropna = False)
```

```python
# Reset pandas view rows
pd.reset_option('display.max_rows')
```

# 05. Basic Customer Statistics using Flag

```python
# Use aggregate function on loyalty flag to find mean, min, max
df_ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean', 'min', 'max']})
```

Based on the loyalty flag all customers spend about the same. The just a couple of pennies difference between the different loyalty flag groups.

# 06. User item Spending Flag

```python
# Create a column for average price each user pays per item
df_ords_prods_merge['average_spend_item'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Assign Spender Flag based on average item price per user
df_ords_prods_merge.loc[df_ords_prods_merge['average_spend_item'] < 10, 'spnd_itm_price_flag'] = 'Lwr Itm Price Spndr'
df_ords_prods_merge.loc[df_ords_prods_merge['average_spend_item'] >= 10, 'spnd_itm_price_flag' ] = 'Hi Itm Price Spndr'
```

```python
# Check Spender Flag
df_ords_prods_merge['spnd_itm_price_flag'].value_counts(dropna = False)
```

```python
#View more rows when using head
pd.set_option('display.max_rows', 100)
```

```python
df_ords_prods_merge[['user_id', 'average_spend_item', 'spnd_itm_price_flag']].head(100)
```

```python
# Drop duplicate user_id to properly check spender flag
df_spender_item_check = df_ords_prods_merge[['user_id', 'average_spend_item', 'spnd_itm_price_flag']].drop_duplicates(subset = 'user_id')
```

```python
# Check spender flag using new column average_spender with no duplicate user_id
df_spender_item_check.head(100)
```

```python
# Reset pandas view rows
pd.reset_option('display.max_rows')
```

```python
# Check Spender Flag
df_spender_item_check['spnd_itm_price_flag'].value_counts(dropna = False)
```

I believe this is a good start to show how much the users pay on averrage for items but its does not actually show us if the user is a high or low spender.

# 07. User Order Spending Flag

```python
# Create a column for average price each user pays per order
df_ords_prods_merge['total_order_spend'] = df_ords_prods_merge.groupby(['order_id'])['prices'].transform('sum')
```

```python
# Assign Spender Flag based on average item price per user
df_ords_prods_merge.loc[df_ords_prods_merge['total_order_spend'] < 50, 'order_spend_flag'] = 'Lower Spender'
df_ords_prods_merge.loc[df_ords_prods_merge['total_order_spend'] >= 50, 'order_spend_flag' ] = 'Higher Spender'
```

```python
# Check Order Spender Flag
df_ords_prods_merge['order_spend_flag'].value_counts(dropna = False)
```

```python
#View more rows when using head
pd.set_option('display.max_rows', 100)
```

```python
df_ords_prods_merge[['user_id', 'order_id', 'total_order_spend', 'order_spend_flag']].head(100)
```

```python
# Drop duplicate user_id to properly check spender flag
df_order_spend_check = df_ords_prods_merge[['user_id', 'order_id', 'total_order_spend', 'order_spend_flag']].drop_duplicates(subset = 'order_id')
```

```python
#View more rows when using head
pd.set_option('display.max_rows', 100)
```

```python
# Check order spend flag using new column total_order_spend with no duplicate user_id
df_order_spend_check.head(100)
```

```python
# Check Order Spender Flag in df_order_spend_check dataframe
df_order_spend_check['order_spend_flag'].value_counts(dropna = False)
```

```python
# Reset pandas view rows
pd.reset_option('display.max_rows')
```

I rather use what the user has spent on average per order instead of the average spent on items as it reflects better the users spending habits. The average price of items users get might be more of an indication of what items actually cost rather than what the user is willing to spend.

# 08. User Frequency Flag

```python
df_ords_prods_merge['customer_freq'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Assign a frequency Flag based on regularity of users ordering
df_ords_prods_merge.loc[df_ords_prods_merge['customer_freq'] > 20, 'customer_freq_flag'] = 'Frequent Customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['customer_freq'] > 10) & (df_ords_prods_merge['customer_freq'] <= 20), 'customer_freq_flag'] = 'Regular Customer'
df_ords_prods_merge.loc[df_ords_prods_merge['customer_freq'] <= 10, 'customer_freq_flag'] = 'Non-Frequent Customer'
```

```python
# Check frequency Flag
df_ords_prods_merge['customer_freq_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.head()
```

```python
# Drop duplicate user_id per order to properly check frequency flag
df_freq_check = df_ords_prods_merge[['user_id', 'customer_freq', 'customer_freq_flag']].drop_duplicates(subset = 'user_id')
```

```python
#View more rows when using head
pd.set_option('display.max_rows', 100)
```

```python
# Check frequency flag with no duplicate orders per user
df_freq_check.head(100)
```

```python
df_freq_check['customer_freq_flag'].value_counts(dropna = False)
```

```python
# Reset pandas view rows
pd.reset_option('display.max_rows')
```
