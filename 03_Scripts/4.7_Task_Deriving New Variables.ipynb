# 04. Creating a Defined Function

```python
# Creating a price Flag
def price_label(row):
    if row['prices'] <= 5: 
       return 'Low-range product'
    elif(row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
# Testind a defined function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Count the defined function output
df['price_range'].value_counts(dropna = False)
```

```python
# Try max function
df['prices'].max()
```

# 05. If-statements with For-loops

```python
# Count the orders per day of the week to determine frequency
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# For-loop to assign a string value based on how busy the day is
result = []

for value in df_ords_prods_merge['orders_day_of_week']:
  if value == 0:
    result.append('Busiest day')
  elif value == 4:
    result.append('Least busy')
  else:
    result.append('Regularly busy')
```

```python
# Print Result
result
```

```python
# Create a new column that displays the frequncy string value
df_ords_prods_merge['daily_frequency'] = result
```

```python
# Count daily_frequency column
df_ords_prods_merge['daily_frequency'].value_counts(dropna = False)
```

```python
#Print orders_day_of_week and daily_frequency columns
print(df_ords_prods_merge[['orders_day_of_week', 'daily_frequency']])
```

# 06. Rewrite If-statements with For-loops

```python
# Adding client request of having two busiest days and two slowest days instead of one
result_new = []

for value in df_ords_prods_merge['orders_day_of_week']:
  if value in [0, 1]:
    result_new.append('Busiest days')
  elif value in [3, 4]:
    result_new.append('Slowest days')
  else:
    result_new.append('Regularly days')
```

```python
# Create a new column that displays the frequncy string value of result_new
df_ords_prods_merge['new_daily_frequency'] = result_new
```

```python
# Count new_daily_frequency
df_ords_prods_merge['new_daily_frequency'].value_counts(dropna = False)
```

Observations on new_daily_frequency Column.
I can see that the regular days got reduced and both the Busiest days and Slowest days increased in count. Based on the original column the distribution looks as expected.

```python
# Count the orders per hour of day to determine frequency by hour
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# For-loop to assign a string value based on how busy the hours of the day are
result_ohod = []

for value in df_ords_prods_merge['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13, 12]:
    result_ohod.append('Most Orders')
  elif value in [3, 4, 2, 5, 1, 0]:
    result_ohod.append('Fewest Orders')
  else:
    result_ohod.append('Average Orders')
```

```python
# Create a new column that displays the frequncy string value of result_ohod
df_ords_prods_merge['busiest_period_of_day'] = result_ohod
```

```python
# Count the period of the day to determine frequency
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
